```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
df_train = pd.read_csv('SalaryData_Train.csv')
```

```python
df_test = pd.read_csv('SalaryData_Test.csv')
```

```python
df_train.head()
```

```python
df_train.drop(labels=['maritalstatus','relationship','race'],axis=1,inplace=True)
```

```python
df_test.drop(labels=['maritalstatus','relationship','race'],axis=1,inplace=True)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
inputs = df_train.drop(labels='Salary',axis=1)
```

```python
target = df_train['Salary']
```

```python
inputs_test = df_test.drop(labels='Salary',axis=1)
```

```python
target_test = df_test['Salary']
```

```python
dummies = pd.get_dummies(data =inputs ,columns= ['workclass','education','occupation','sex','native'])
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis=1)
```

```python
inputs.columns
```

```python
inputs.drop(columns=['workclass','education','occupation','sex','native','workclass_ Federal-gov','education_ 10th',
                     'occupation_ Armed-Forces','sex_ Female','native_ Cambodia'],axis=1,inplace=True)
```

```python
inputs.isna().sum()
```

```python
dummies_test = pd.get_dummies(data =inputs_test ,columns= ['workclass','education','occupation','sex','native'])
dummies_test.head(3)
```

```python
inputs_test = pd.concat([inputs_test,dummies_test],axis=1)
```

```python
inputs_test.drop(columns=['workclass','education','occupation','sex','native','workclass_ Federal-gov','education_ 10th',
                     'occupation_ Armed-Forces','sex_ Female','native_ Cambodia'],axis=1,inplace=True)
```

```python
inputs_test.isna().sum()
```

```python
model = GaussianNB()
```

```python
model.fit(inputs,target)
```

```python
model.score(inputs_test,target_test)
```

```python
model.score(inputs,target)
```
